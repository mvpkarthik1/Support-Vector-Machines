```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
SalaryData_Train = pd.read_csv("SalaryData_Train(1).csv")
SalaryData_Train.info()
```

```python
SalaryData_Train
```

```python
SalaryData_Test = pd.read_csv("SalaryData_Test(1).csv")
SalaryData_Test.info()
```

```python
SalaryData_Test
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
categ=['age','workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary']
SalaryData_Train[categ]=SalaryData_Train[categ].apply(labelencoder.fit_transform)
SalaryData_Train
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
categ=['age','workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary']
SalaryData_Test[categ]=SalaryData_Test[categ].apply(labelencoder.fit_transform)
SalaryData_Test
```

```python
SalaryData = pd.concat([SalaryData_Train,SalaryData_Test],axis=0)
SalaryData
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,30))
sns.heatmap(SalaryData.corr(),annot=True)
```

```python
##Normalising the data as there is scale difference
predictors = SalaryData.iloc[:,0:13]
target = SalaryData.Salary

def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)

fires = norm_func(predictors)
fires
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size = 0.25, stratify = target)
```

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(x_train,y_train)
pred_test_linear = model_linear.predict(x_test)

np.mean(pred_test_linear==y_test)
```

```python
# Kernel = poly
model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
pred_test_poly = model_poly.predict(x_test)

np.mean(pred_test_poly==y_test)
```

```python
# kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
pred_test_rbf = model_rbf.predict(x_test)

np.mean(pred_test_rbf==y_test) #Accuracy = 74.6%
```

```python

#'sigmoid'
model_sig = SVC(kernel = "sigmoid")
model_sig.fit(x_train,y_train)
pred_test_sig = model_rbf.predict(x_test)

np.mean(pred_test_sig==y_test) #Accuracy = 73%
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
clf = SVC()
param_grid = [{'kernel':['rbf'],'C':[15,14,13,12,11,10,1] }]
gsv = GridSearchCV(clf,param_grid)
gsv.fit(x_train,y_train)
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score


clf = SVC(kernel ="poly", C= 15, gamma = 50)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
